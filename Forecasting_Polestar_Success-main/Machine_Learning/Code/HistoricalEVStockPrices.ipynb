```python
#Input Ticker to test
ticker="LCID"
```

```python
tsla = yf.Ticker(f"{ticker}")
tsla_hist = tsla.history(period="max")
```

```python
#Display tthe data
tsla_hist.head(10)
```

```python
tsla_hist_df = tsla_hist
tsla_hist_df
```

## Data Cleaning 

```python
#Reset Index 
tsla_hist_df = tsla_hist_df.reset_index()
tsla_hist_df
```

```python
tsla_hist_df=tsla_hist_df[["Date", "Close", "Volume"]]
tsla_hist_df
```

```python
tsla_hist_df["Ticker"]=f"{ticker}"
tsla_hist_df
```

```python
filename=f"{ticker}.csv"
```

```python
tsla_hist_df.to_csv(filename, sep=',', encoding='utf-8', index=False)
```
